```python
winemag_dataset_points_mean = winemag_dataset.points.mean()
winemag_dataset.points.map(lambda p: p - winemag_dataset_points_mean)
```

```python
winemag_dataset.country + " - " + winemag_dataset.region_1
```

```python
def remean_points(row):
    row.points = row.points - winemag_dataset_points_mean
    return row

winemag_dataset.apply(remean_points, axis='columns')
```
